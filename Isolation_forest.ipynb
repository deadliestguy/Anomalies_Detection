## outliers find korte hbe
## anomaly detection -> isolation forest 

```python
from sklearn.ensemble import IsolationForest

# model creation 
model = IsolationForest(contamination=0.20)  # contamination = threshold value 

# train the model 
model.fit(df)

# predict 
prediction = model.predict(df)
```

```python
# prediction = -1 ->outliers , 1= normal point 
```

```python
prediction
```

```python
# -1 er index alada korbo 
index = np.where(prediction<0)
index
```
