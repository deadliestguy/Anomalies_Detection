```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X,y = make_circles(n_samples=450,noise=0.1,factor=0.3)
```

```python
# plot teh dataaset 
plt.scatter(X[:,0],X[:,1])
```

```python
# dbscan 
from sklearn.cluster import DBSCAN

# model creation 
clf = DBSCAN(eps=0.10,min_samples=4)

prediction = clf.fit_predict(X) 
prediction
```

```python
# plot teh dataaset 
plt.scatter(X[:,0],X[:,1] , c=prediction)
```
