## Local Outlier Factor 

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X,y = make_circles(n_samples=450,noise=0.1,factor=0.3)
```

```python
# plot teh dataaset 
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=10,algorithm='auto',)
```

```python
# fit the model 
prediction = lof.fit_predict(X)
prediction
```

```python
# plot teh dataaset 
plt.scatter(X[:,0],X[:,1],c=prediction)
```
